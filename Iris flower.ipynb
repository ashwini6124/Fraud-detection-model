```python
iris_df = load_iris()
X = iris_df.data 
Y = iris_df.target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
     X, Y, test_size=0.4, random_state=20)
```

```python
model = LogisticRegression(max_iter=200)
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
print("Predicted lables:", y_pred)
```
